```python
import pandas as pd
import numpy as np
# originally designed to parse https://opendata.arcgis.com/datasets/496533836db640bcade61dd9078b0d63_53.csv
csv_filename = 'path/to/some_csv_file.csv'
df = pd.read_csv(csv_filename)
#df.PROPTYPE.unique()
#df.stack().unique()

mes = '\n' + '===  DISCARDED  ==='
print(mes)
for c in pd.unique(df.keys()):
    distinct_list = df[c].unique()
    mes = ' ==> ' + c
    if len(distinct_list) == 1 and str(distinct_list[0]) == 'nan':
        mes = c + ' ==> ' + str(distinct_list[0])
        print(mes)
        # print(distinct_list)

mes = '\n' + '===  UNIQUE VALUES  ==='
print(mes)
for c in pd.unique(df.keys()):
    distinct_list = df[c].unique()
    mes = ' ==> ' + c
    if len(distinct_list) == 1 and str(distinct_list[0]) != 'nan':
        mes = c + ' ==> ' + str(distinct_list[0])
        print(mes)
        # print(distinct_list)

mes = '\n' + '===  LIMITED LIST OF VALUES  ==='
print(mes)
for c in pd.unique(df.keys()):
    distinct_list = df[c].unique()
    mes = ' ==> ' + c
    if 1 < len(distinct_list) <= 20:
        mes = ' ==> ' + c
        print(mes)
        #print(distinct_list)

mes = '\n' + '===  EXTENDED LIST OF VALUES  ==='
print(mes)
for c in pd.unique(df.keys()):
    distinct_list = df[c].unique()
    mes = ' ==> ' + c
    if len(distinct_list) > 20:
        mes = ' ==> ' + c
        print(mes)
        print(distinct_list)

```
